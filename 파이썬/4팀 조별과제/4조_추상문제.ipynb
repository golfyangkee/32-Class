### a_bookclass.py

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_checked_out = False

    def check_in(self):
        self.is_checked_out = False
    def check_out(self):
        self.is_checked_out = True

    def __str__(self): #책의 리스트와 대여 상태를 형태에 맞춰 출력합니다.
        #self.is_checked_out이 True인 경우 "대출 중 " 문자열 할당
        checked = "대출 중" if self.is_checked_out else "대여 가능"
        return f"[{checked}] {self.title} - {self.author}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def check_in_book(self, title):
        for book in self.books:
            if book.title == title and book.is_checked_out:
                book.check_in()
                return

    def check_out_book(self, title):
        for book in self.books:
            if book.title == title and not book.is_checked_out:
                book.check_out()
                return

    def find_by_author(self, author):
        return [book for book in self.books if book.author == author]

    def find_by_title(self, title):
        return [book for book in self.books if book.title == title]

    def is_available(self, title):
        for book in self.books:
            if book.title == title:
                return not book.is_checked_out
        return False

    def remove_book(self, title):
        self.books = [book for book in self.books if book.title != title]

    def __str__(self):
        return "\n".join(str(book) for book in self.books)


if __name__ == "__main__":
    library = Library()
    book1 = Book("위대한 개츠비", "F. 스콧 피츠제럴드")
    book2 = Book("1984", "조지 오웰")

    library.add_book(book1)
    library.add_book(book2)

    book_to_checkout = "위대한 개츠비"
    if library.is_available(book_to_checkout):
        library.check_out_book(book_to_checkout)
        print(f"'{book_to_checkout}'을(를) 대출하였습니다.")
    else:
        print(f"'{book_to_checkout}'은(는) 대출 가능한 상태가 아닙니다.")

    print("\n도서관 현황:")
    print(library)

    library.check_in_book(book_to_checkout)
    print(f"\n'{book_to_checkout}'을(를) 반납하였습니다.")


    print("\n도서관 현황:")
    print(library)
```

### b_abstractMobile.py

```python
from abc import abstractmethod, ABCMeta


class IChange(metaclass=ABCMeta):

    @abstractmethod
    def charge(self):
        pass

class Mobile(metaclass=ABCMeta):

    @abstractmethod
    def __init__(self,mobileName,batterySize,osType):
        self.mobileName = mobileName
        self.batterySize = batterySize
        self.osType      = osType

    @abstractmethod
    def operate(self):
        pass

    @abstractmethod
    def __str__(self):
        pass


class LTab(IChange,Mobile):
    def __init__(self,mobileName,batterySize,osType):
            super(LTab, self).__init__(mobileName,batterySize,osType)

    @property
    def SetbatterySize(self):
        return self.batterySize

    @SetbatterySize.setter
    def SetbatterySize(self,b):
        if b < 0:
            print('0보다 작은 숫자는 배터리로 불가합니다')
        else:
            self.batterySize = b
    def operate(self,minute):
        self.batterySize -= (10 * minute)
        if self.batterySize < 0:
            self.batterySize = 0

    def charge(self,minute):
        self.batterySize +=  (10 * minute)
        if self.batterySize > 100:
            self.batterySize = 100



    def __str__(self):
        return f'{self.mobileName}\t\t{self.batterySize}\t\t{self.osType}'


class OTab(IChange,Mobile):
    def __init__(self,mobileName,batterySize,osType):
            super(OTab, self).__init__(mobileName,batterySize,osType)

    @property
    def SetbatterySize(self):
        return self.batterySize

    @SetbatterySize.setter
    def SetbatterySize(self, b:int):
        if b < 0:
            print('0보다 작은 숫자는 배터리는 불가합니다')
        else:
            self.batterySize = b

    def operate(self,minute):
        self.batterySize -= (12 * minute)
        if self.batterySize < 0:
            self.batterySize = 0

    def charge(self,minute):
        self.batterySize += (8 * minute)
        if self.batterySize > 100:
            self.batterySize = 100

    def __str__(self):
        return f'{self.mobileName}\t\t{self.batterySize}\t\t{self.osType}'



if __name__ == '__main__':
    lt = LTab("LTab", 50, "AP-01")
    ot = OTab("OTab", 100, "AND-20")
    
    # SetbatterySize 테스트
    #lt.SetbatterySize = 60

    # 출력 함수
    def print_status(title, devices):
        print(title)
        print("Mobile\t\tBattery\t\tOS")
        print("---------------------------------------------")
        for device in devices:
            print(device)
        print()


    # 상태 출력
    devices = [lt, ot]
    print_status("초기 상태", devices)

    # 10분 충전 후 상태 출력
    for device in devices:
        device.charge(10)
    print_status("10분 충전", devices)

    # 5분 사용 후 상태 출력
    for device in devices:
        device.operate(5)
    print_status("5분 통화", devices)
```

### c_student.py

```python
# Student 클래스와 StudentManager 클래스를 작성합니다.
#모든 메소드를 채우세요

class Student:
    def __init__(self, name, student_id, score):
        self.name = name
        self.student_id  = student_id
        self.score = score

    def update_score(self, new_score):
        self.score = new_score

    def __repr__(self):
        return f'{self.name} (ID : {self.student_id} 점수: {self.score})'


class StudentManager:
    def __init__(self):
        self.Students=[]

    def add_student(self, student):
                self.Students.append(student)

    def remove_student(self, student_id):
        for Student in self.Students:
            if Student.student_id == student_id:
                self.Students.remove(Student)

    def get_students_with_score_above(self, threshold):
            parthres = threshold
            aboveList = [a for a in self.Students if a.score >= parthres]
            return aboveList


    def get_top_students(self, number_of_students): #상위 학생 목록 조회
        sorted_students = sorted(self.students, key=lambda x: x.score, reverse=True)
        return sorted_students[0:number_of_students]
        ## sorted(iterable_정렬대상, key=key_정렬기준, reverse=reverse_정렬순서T=내림차,F=오름차)
        ## 64행에서 마지막 숫자(2)를 받아 상위 학생 2명을 내림차순 정렬합니다
        
    def __str__(self):
        res = ''
        for student in self.Students:
            res += f'{student.name} (ID : {student.student_id} 점수: {student.score})\n'
        return res


if __name__ == "__main__":
    # 학생 관리자 객체 생성 및 학생 목록 초기화
    manager = StudentManager()

    # 학생 추가
    manager.add_student(Student("홍길동", "2021001", 85))
    manager.add_student(Student("김철수", "2021002", 92))
    manager.add_student(Student("이영희", "2021003", 76))

    # 특정 점수 이상 학생 목록 조회
    students_above_80 = manager.get_students_with_score_above(80)
    print("점수가 80 이상인 학생들:")
    for student in students_above_80:
        print(student)

    # 상위 학생 목록 조회
    top_students = manager.get_top_students(2)
    print("\n상위 학생 목록:")
    for student in top_students:
        print(student)

    # 학생 관리자 상태 출력
    print("\n학생 관리자 상태:")
    print(manager)
```

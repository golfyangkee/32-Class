```python
from sklearn.datasets import load_iris
from sklearn.tree import plot_tree, DecisionTreeClassifier
import matplotlib.pyplot as plt

iris=load_iris()
clf = DecisionTreeClassifier().fit(iris.data, iris.target)

plt.figure(figsize=(12,9))
plot_tree(clf, filled=True)
plt.title("Decision tree trained on all the iris features")
plt.show()

# 분할 기준, 샘플수, 불순도, 클래스 간의 분포를 설명 -> 그래프 확인
# gini= 불순도
# 불순도를 최소화 하는 방향으로 데이터를 분할한다. -> 불할 기준
# X가 가지고 있는 2인데 X의 2번째 특징을 얘기한다. X의 2번지 0 1 2 3번째 X 변수를 기점으로 분할을 하겠다.
# 이 특성의 값이 2.45 이하이냐? 이하일 경우
# 얘가 가진 gini의 불순도는 0.667이다
# 여기가 yes or no 니까
# 첫번째 박스는 0,1,2 분류가 50씩 분류된거니까 X[2]가 꽃잎의 길이니까 꽃잎이 2.45cm 이하냐? 라고 물어보는 것
# T는 왼쪽 F는 오른쪽인데 잘 못 하면 바뀔수도 있다.
```

```python
from sklearn.preprocessing import LabelEncoder

iris  = load_iris() 
X,y  = iris.data[50:,[1,2]] , iris.target[50:]

le= LabelEncoder() # -> 문자일 경우 y값을 정수로 변환, 클래스가 0, 1, 2일때 0, 1로만 분류하고 싶을 때, 혹은 str를 정수로 변환하고 싶을 때
res=le.fit_transform(y)
res, le.classes_
```

```python
import numpy as np
# 타겟 변수를 문자열로 선언 후 변환해 보자.

y_str = np.array(["Class1" if label == 1 else "Class2" for label in y]) # y 가 수치라서 일부로 str로 바꿔주고

le= LabelEncoder() # -> y값을 정수로 변환
res=le.fit_transform(y_str)

y_str, res, le.classes_
```

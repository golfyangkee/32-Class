### 지역변수 & 전역변수(Global)
- 변수는 자신이 생성된 범위(코드 블럭) 안에서만 유효하다.
- 함수 안에서 만든 변수는 함수 안에서 살아 있다가 함수 코드의 실행이 종료되면 소멸된다. -> 지역변수
- 함수 외부에서 만든 변수는 프로그램이 살아있는 동안에 함께 살다가 프로그램이 종료되면 같이 소멸된다. -> 전역변수
- 전역변수를 사용하는 경우는?  프로그램에서 공통적으로 사용되고 잘 변하지 않는 데이터(정적데이터)를 전역변수로 선언한다.


```python
strdata='전역변수'  # 함수 외부에 선언된 변수

def func1():
    strdata2='지역변수' # 함수 내부에 선언된 변수
    return strdata2

func1() # 지역변수
```

```python
pi=3.141592   # 함수 외부에 선언된 전역변수  (정적데이터 변수 = 공유변수)

def cycle_fun1(num1): # 원의 넓이를 구하는 함수
    global pi   # global 이란 키워드를 앞에 선언한다. 반드시 명시 해야한다.
    return pi*num1*num1

def cycle_fun2(num1): # 원의 1/4인 부채꼴의 넓이를 구하는 함수
    global pi   # global 이란 키워드를 앞에 선언한다.
    return pi*num1*num1/4

print('원의 넓이: ',cycle_fun1(5))
print('원의 1/4 넓이: ',cycle_fun2(5))
```

```python
def my_abs (su):  # 만일에 su가 0보다 크면 리턴하고 그렇지 않으면 *-1을 해서 리턴하자
    if su<0:             # su가 0보다 작으면
        return su*(-1)   # -1을 곱해서 리턴
    else:                # 그렇지 않으면
        return su        # 입력값 그대로 리턴
```

## array object 설명
- "Array"는 프로그래밍에서 데이터를 저장하는 데 사용되는 구조로, 동일한 데이터 타입의 요소들을 순차적으로 나열한 것을 말합니다. "Array"는 많은 프로그래밍 언어에서 기본적인 데이터 구조 중 하나입니다. 넘파이(Numpy)에서는 다차원 배열(numpy.ndarray)이라는 자료형이 제공되어 배열을 효과적으로 다룰 수 있습니다.

### array object 특징
- 고정된 크기: 대부분의 array는 생성할 때 크기가 고정되며, 이 크기를 변경하기 어렵습니다. 크기를 동적으로 조절할 수 있는 자료형도 있긴 하지만, 대부분의 array는 초기에 지정된 크기를 유지합니다.

- 같은 데이터 타입: Array 내의 요소들은 동일한 데이터 타입을 가지고 있어야 합니다. 이로써 각 요소의 크기를 일정하게 유지하고 메모리를 효율적으로 사용할 수 있습니다.

- 인덱스로 접근: 각 요소는 배열에 저장된 위치(인덱스)로 접근됩니다. 대부분의 프로그래밍 언어에서 인덱스는 0부터 시작하며, 순차적으로 증가합니다.

- 빠른 접근 및 검색: 배열은 인덱스를 사용하여 특정 요소에 빠르게 접근할 수 있습니다. 이는 메모리에 연속적으로 저장되어 있기 때문에 가능합니다.

- 다차원 배열: 몇몇 언어와 라이브러리는 다차원 배열도 지원합니다. 1차원 배열은 벡터, 2차원 배열은 행렬, 그 이상의 차원은 텐서 등으로 불립니다.

### np.flags
- ndarray.flags 속성은 NumPy 배열(ndarray)의 속성과 관련된 여러 플래그(Flags) 정보를 포함하는 속성입니다. 이 속성을 통해 배열이 어떻게 저장되고 메모리에 배치되어 있는지에 대한 정보를 확인할 수 있습니다.

#### 일반적으로 ndarray.flags 속성에서 접근할 수 있는 중요한 플래그들은 다음과 같습니다:

- C_CONTIGUOUS (CORDER): 배열이 C 스타일의 연속된 메모리 레이아웃을 가지고 있는지 여부를 나타내는 플래그입니다. 만약 True이면 C 스타일의 연속된 레이아웃을 갖습니다.

- F_CONTIGUOUS (FORDER): 배열이 Fortran 스타일의 연속된 메모리 레이아웃을 가지고 있는지 여부를 나타내는 플래그입니다. 만약 True이면 Fortran 스타일의 연속된 레이아웃을 갖습니다.

- OWNDATA: 배열이 메모리를 소유하고 있는지 여부를 나타내는 플래그입니다. 만약 True이면 배열이 메모리를 소유하고 있습니다.

- WRITEABLE (WRITEBACKIFCOPY): 배열이 쓰기 가능한지 여부를 나타내는 플래그입니다. 만약 False이면 읽기 전용입니다.

- ALIGNED: 배열의 메모리가 특정 바이트 경계에 정렬되어 있는지 여부를 나타내는 플래그입니다.

### 확률밀도함수 pdf()
- 확률밀도함수(Probability Density Function, PDF)는 확률 변수의 값이 어떤 구간에 속할 확률을 나타내는 함수입니다. 확률 변수가 특정 값에 대해 어떤 값이냐보다 어떤 구간에 속할 확률에 관심이 있는 경우 사용됩니다.
- 확률밀도함수는 연속확률변수(continuous random variable)에 사용되며, 이 함수를 통해 확률 변수의 확률 분포를 표현합니다. 이 함수는 주로 수학적 모델링, 통계 분석, 확률 이론 등 다양한 분야에서 사용됩니다.
- pdf(x, loc=0, scale=1)

#### 특성
1) 비음수성(Non-negativity): 확률밀도함수는 모든 가능한 값에 대해 0 이상의 값을 가져야 합니다. 즉, f(x)≥0이어야 합니다.
2) 정규화(Normalization): 전체 확률 공간에서의 확률의 합은 1이어야 합니다.
3) 구간 확률 계산: 확률밀도함수를 사용하면 확률 변수가 어떤 구간에 속할 확률을 적분을 통해 계산할 수 있습니다.
4) 무한히 많은 가능한 값: 확률밀도함수는 확률 변수가 취할 수 있는 모든 가능한 값에 대해 정의되어야 하며, 무한히 많은 가능한 값이 있을 수 있습니다.

```python
# pdf() -> 연속확률 변수의 확률 분포를 설명한다. // 확률밀도 함수
# 확률 밀도 함수는 음수가 아니어야 한다. 적분값(전체확률) = 1 반드시 1이어야 한다.
# 정규분포(가우스분포), 지수분포, 균일분포, 베타분포 등을 매개변수 (평균, 표준편차, 형상모수, 척도모수)에 따라 
# 그래프를 다르게 그려준다.

import numpy as np
from scipy.stats import norm # scipy 의 stats 패키지로 부터
                             # norm 이라는 모듈을 import 해라
# from 패키지 import 모듈이름

avg=148.5
std = 30
N=1000000
height = np.random.randn(N)*std+avg
a=[]

for i in range(1,10001):
    result= np.random.choice(height,100).mean()  # 모집단에서 100개를 추출해서 평균값 취함
    a.append(result)   # 평균값을 10000번 입력함
    
x=np.arange(140,160,0.001)    
# 140~160까지 0.001 간격으로 숫자를 생성 // 1mm 간격으로 나누어서 이거 가져다가 확률분포로 보겠다는 의미
y=norm.pdf(x,np.mean(a), np.std(a))  # 초등학생 키의 표본평균값들에 대한 확률 밀도함수 값이 출력
print(y)
```
